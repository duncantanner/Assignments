```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 1, 50)

y_log = np.log(x)
z_exp = np.exp(x)

plt.scatter(x, y_log, label = 'Natural Log')
plt.scatter(x, z_exp, label = 'Exponetial')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Natural Log and Exponential Functions')
plt.legend(loc = 'lower right')
plt.show()

y = np.linspace(-6.5, 6.5, 100)

y_sin = np.sin(y)
z_cos = np.cos(y)

plt.plot(y, y_sin, label = 'Sin')
plt.plot(y, z_cos, label = 'Cos')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Sin and Cos')
plt.legend(loc = 'lower left')
```
